```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plot
%matplotlib inline

# FIXME: Tweak learning rates, dropouts, batchnormalization, regularlization 
```

```python
redwine_data = pd.read_csv(r'redwine_data.csv', sep=',', index_col=0)

```

```python
redwine_data.head()
```

```python
# Data proprocessing
features = redwine_data.iloc[:,0:11].values
labels = redwine_data.iloc[:,11].values
```

```python
print(features)
```

```python
# convert labe array to one hot encoded matrix

from sklearn.preprocessing import LabelEncoder
encoder = LabelEncoder()
y = encoder.fit_transform(labels)
labels = pd.get_dummies(y).values
```

```python
# split data into training and test sets

from sklearn.model_selection import train_test_split
train_features, test_features, train_labels, test_labels = train_test_split(features, labels, test_size = 0.2, random_state = 0)
```

```python
#scaling the data
from sklearn.preprocessing import StandardScaler
feature_scaler = StandardScaler()
train_features = feature_scaler.fit_transform(train_features)
test_features = feature_scaler.transform(test_features)
```

```python
import keras
from keras.models import Sequential
from keras.layers import Dense
```

```python
# Initialize the network
ann_clf = Sequential()
```

```python
# adding input and hidden layers

#TODO: Try better initializer. e.g
#keras.initializers.he_normal(seed=None)

# hidden layer
# units = average of input and output nodes, (11 input nodes + 6 output nodes) / 2  = 8.5
# relu = rectifier function
h1 = Dense(units = 9, kernel_initializer='he_normal', activation='relu', input_dim=11)
ann_clf.add(h1)

# second hidden layer
h2 = Dense(units = 9, kernel_initializer = 'he_normal', activation='relu')
ann_clf.add(h2)
```

```python
# output layers
op = Dense(units=6, kernel_initializer = 'he_normal', activation='softmax')
ann_clf.add(op)
```

```python
# Training the neural network

# Configure the right optimizer and loss functions
ann_clf.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])

```

```python
# Start training
ann_clf.fit(train_features, train_labels, batch_size=1, epochs=500)
```

```python
predictions=ann_clf.predict(test_features)
predictions=np.argmax(predictions, axis=1)
test_labels=np.argmax(test_labels,axis=1)
```

```python
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
print(confusion_matrix(test_labels,predictions))
print(classification_report(test_labels, predictions))
print(accuracy_score(test_labels, predictions))
```
